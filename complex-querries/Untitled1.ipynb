```python
con = sql.connect('chinook.db')
```

```python
def run(q):
    return pd.read_sql_query(q,con)
```

# WORKING ON COUNTRY==USA INSTEAD OF STATE==NEW_YORK

```python
q = """
select first_name || " " || last_name as Customers,
state, country, sum(unit_price) as Total_Spend
from customer
inner join invoice
on customer.customer_id == invoice.customer_id
inner join invoice_line
on invoice.invoice_id == invoice_line.invoice_id
where country == 'USA'
group by customer.customer_id
"""
run(q)
```

```python
q = """
select first_name || " " || last_name as Customers,
state, country, sum(unit_price) as Total_Spend
from customer
inner join invoice
on customer.customer_id == invoice.customer_id
inner join invoice_line
on invoice.invoice_id == invoice_line.invoice_id
where country == 'USA'
group by customer.customer_id
having sum(unit_price) > 80
"""
run(q)
```

```python
q = """
select first_name || " " || last_name as Customers,
state as State, country as Country, count(invoice_id) as No_Of_Visits, sum(total) as Total_Spend

from customer
inner join invoice
on customer.customer_id == invoice.customer_id

where customer.country == 'USA'
group by customer.customer_id
having Total_Spend > 80
"""
run(q)
```

```python
q = """
select invoice_line_id,invoice.invoice_id
from invoice
inner join invoice_line
on invoice.invoice_id == invoice_line.invoice_id

where invoice.billing_country == 'USA'
group by invoice.invoice_id
limit 5
"""
run(q)
```

```python
q = """
select first_name || " " || last_name as Customers,
state as State, country as Country, count(invoice_line.invoice_id) as No_Of_Visits,
sum(invoice_line.unit_price) as Total_Spend

from customer
inner join invoice
on customer.customer_id == invoice.customer_id
inner join invoice_line
on invoice.invoice_id == invoice_line.invoice_id


where customer.country == 'USA'
group by customer.customer_id
having Total_Spend > 80
"""
run(q)
```
